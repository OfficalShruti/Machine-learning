```python

#[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
#['Action','Adventure','FFantasy','SciFi']<- in this we want to convert
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

# New Section

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
  L=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      L.append(i['name'])
      counter+=1
    else:
      break
  return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      L.append(i['name'])
      break
  return L
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#remove space btween the name and surname
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']= movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new_df=movies[['movie_id', 'title', 'tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
import nltk
!pip install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append (ps.stem(i))
  return" ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
vectors[0] #first movie avtar
```

```python
cv.get_feature_names()
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver')
```

```python
from sklearn.metrics.pairwise import cosine_similarity # 0 to 1 1 means more similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])), reverse= True, key=lambda x:x[1])[1:6]#x:x[1 0] y cloumne 2 k liye nikaleaga 0 index column h 1st similarity column to uske basise pe nikalega
```

```python
def recommend(movie):
  movie_index =  new_df[new_df['title']==movie].index[0]
  distances=similarity[movie_index]
  movies_list=sorted(list(enumerate(distances)), reverse= True, key=lambda x:x[1])[1:6]

  for i in movies_list:
    print(new_df.iloc[i[0]].title)
  return
```

```python
recommend('Batman Begins')
```
